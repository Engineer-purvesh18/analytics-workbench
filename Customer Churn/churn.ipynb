DATA CLEANING

```python
# creating a copy of df for data manipulation so that the actual data isnt used.

df1 = df.copy()
```

```python
df1.head()
```

```python
# Converting the TotalCharges from str to numeric
df1.TotalCharges = pd.to_numeric(df1.TotalCharges, errors='coerce')
df1.isnull().sum()
```

```python
# Looking for null values and plotting them if any
missing = pd.DataFrame((df1.isnull().sum())*100/df1.shape[0]).reset_index()
plt.figure
ax = sns.pointplot(x='index', y=0, data=missing)
plt.xticks(rotation = 90, fontsize  = 7)
plt.title("Percentage of Missing Values")
plt.ylabel("Percentage")
plt.show()
```

```python
df1.loc[df1['TotalCharges'].isnull() == True]
```

Based on the above outputs it can easily be seen that there are 11 missing values in TotalChrages features that is almost 15% of the total number of entries in the feature. 

Since the percentage of missing values is very low, it is safe to drop them 


```python
# dropping the null values
df1.dropna(how='any', inplace = True)
```

```python
df1.isnull().sum()
```

Based on above plot churn happens to be high when the monthly charges are high.

```python
# Comparing Chrun wrt Total Charges using kernel density charges
Tot = sns.kdeplot(new_df_dummies.TotalCharges[(new_df_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(new_df_dummies.TotalCharges[(new_df_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

Churn is high at lower total charges.

```python
def uniplot(df, col, title, hue=None):
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, x=col, hue=hue)
    plt.xticks(rotation=45)
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

CONCLUSION

📊 Customer Churn – Key Observations

Based on the exploratory data analysis, the following patterns and behavioral trends were identified:

1️⃣ Payment Method Insights

Customers using Electronic Check as their payment method exhibit the highest churn rate (~44%).

Female customers paying via Credit Card show relatively higher churn compared to their counterparts using other payment modes.

2️⃣ Contract Type Impact

Customers on Month-to-Month contracts have a significantly higher churn rate (~41%), likely due to the absence of long-term contractual commitment.

Customers with Two-Year contracts demonstrate the lowest churn rates, indicating strong retention associated with longer-term agreements.

3️⃣ Service-Related Factors

Customers without Online Security and Tech Support services are more likely to churn.

The absence of value-added services appears to strongly correlate with higher churn probability.

4️⃣ Demographic Patterns

Senior Citizens show a high churn rate (~42%), indicating potential dissatisfaction or service mismatch within this segment.

Customers without a Partner are more likely to churn (~35%).

Female customers without partners demonstrate a higher churn tendency compared to male customers in the same category.

5️⃣ Tenure & Financial Indicators

Customers with low tenure (1–12 months) are significantly more likely to churn.

Higher Monthly Charges, combined with low tenure and lower Total Charges, are strongly associated with churn.

This suggests early dissatisfaction and potential pricing sensitivity among newer customers.

🎯 Overall Business Implications

Short-term contract customers represent the highest retention risk segment.

Payment method and service subscription patterns provide strong predictive signals.

Early-stage customer experience (first 12 months) is critical for retention.

Bundled service offerings (Security, Tech Support) may improve long-term engagement.
